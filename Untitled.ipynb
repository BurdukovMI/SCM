# Вариант 2. 
Численное вычисление интеграла
$$
\pi = \int_0^1\frac{4}{1+x^2}dx
$$
Нарисуем график этой функции на промежутке [0,1] 


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import integrate
```

```python
xx = np.linspace(0, 1, 500)
yy = 4/(1+xx**2)
plt.plot(xx, yy)
plt.grid()
pass
```

1. Посчитаем интеграл с помощью формулы прямоугольников с 2,4,8,16,32,64,128 узлами.

```python
a = 0
b = 1
for n in [2,4,8,16,32,64,128]:
    h = (b - a)/n
    x = np.arange(a + h/2, b, h)
    y =4/(1+x**2)
    I_rect =h*sum(y)
    plt.plot(n,abs(np.pi-I_rect)/(np.pi),'o',color='tab:orange')
    print('Кол-во узлов ',n,' Погрешность ',abs(np.pi-I_rect)/(np.pi))
Ibest_rect=I_rect
Rectbest_err= abs(np.pi-I_rect)/(np.pi)
pass
Ibest_rect
```

2. Посчитаем интеграл с помощью формулы трапеции с 2,4,8,16,32,64,128 узлами.

```python
for n in [2,4,8,16,32,64,128]:
    h = (b - a)/n
    x = np.linspace(a, b, n + 1)
    y = 4/(1+x**2)
    I_trapz = integrate.trapz(y, x)
    plt.plot(n,(np.pi-I_trapz)/(np.pi),'o',color='tab:blue')
    print('Кол-во узлов ',n,' Погрешность ',(np.pi-I_trapz)/(np.pi))
pass
Trapzbest_err= abs(np.pi-I_trapz)/(np.pi)
Ibest_trapz=I_trapz

```

3. Посчитаем интеграл с помощью правила Симпсона с 2,4,8,16,32,64,128 узлами.

```python
for n in [2,4,8,16,32,64,128]:
    h = (b - a)/n
    x = np.linspace(a, b, n + 1)
    y = 4/(1+x**2)
    I_simp=integrate.simps(y)*h
    plt.plot(n,abs(np.pi-I_simp)/(np.pi),'o',color='tab:red')
    print('Кол-во узлов ',n,' Погрешность ',abs(np.pi-I_simp)/(np.pi))
pass
Simpbest_err= abs(np.pi-I_simp)/(np.pi)
Ibest_simp=I_simp
Ibest_simp
```

```python
print('Rectangaular method ', Ibest_rect,' ',Rectbest_err)
print('Trapezoidal method ', Ibest_trapz,' ',Trapzbest_err)
print('Simpson rule ', Ibest_simp,' ',Simpbest_err)
print('Pi', np.pi)

```

Подведем промежуточный итог:
1.Самым точным ,как и предсказывалось, стал интеграл посчитанный методом Симпсона(так как он имеет 4 порядок сходимости).
2.Стало достаточно неожидано, что интеграл, посчитанный с помощью правилом прямоугольников,стал более точным чем интеграл, посчитанный правилом трапеций. Изначально предсказывалась обратная ситуация, так как 1 способ имеет более маленький порядок сходимости.

# Вариант 3. 
Численное вычисление интеграла
$$
-\frac{4}{9} = \int_0^1 \sqrt{x}\ln x dx
$$
Нарисуем график этой функции на промежутке [0,1] 

```python
xx = np.linspace(1e-9, 1, 500)
yy = np.sqrt(xx)*np.log(xx)
plt.plot(xx, yy)
plt.grid()
pass
```

1. Посчитаем интеграл с помощью формулы прямоугольников с 2,4,8,16,32,64,128 узлами.

```python
a = 1e-12
b = 1
ans=-(4/9)
for n in [2,4,8,16,32,64,128]:
    h = (b - a)/n
    x = np.arange(a + h/2, b, h)
    y =np.sqrt(x)*np.log(x)
    I_rect =h*sum(y)
    plt.plot(n,(ans-I_rect)/abs(ans),'o',color='tab:orange')
    print('Кол-во узлов ',n,' Погрешность ',(ans-I_rect)/abs(ans))
Ibest_rect=I_rect
Rectbest_err= (ans-I_rect)/abs(ans)
pass
Ibest_rect
```

2. Посчитаем интеграл с помощью формулы трапеции с 2,4,8,16,32,64,128 узлами.

```python
for n in [2,4,8,16,32,64,128]:
    x = np.linspace(a, b, n + 1)
    y = np.sqrt(x)*np.log(x)
    I_trapz = integrate.trapz(y, x)
    plt.plot(n,(ans-I_trapz)/(ans),'o',color='tab:blue')
    print('Кол-во узлов ',n,' Погрешность ',(ans-I_trapz)/(ans))
pass
Trapzbest_err= (ans-I_trapz)/(ans)
Ibest_trapz=I_trapz
Ibest_trapz
```

3. Посчитаем интеграл с помощью правила Симпсона с 2,4,8,16,32,64,128 узлами.

```python
for n in [2,4,8,16,32,64,128]:
    h = (b - a)/n
    x = np.linspace(a, b, n + 1)
    y = np.sqrt(x)*np.log(x)
    I_simp=integrate.simps(y)*h
    plt.plot(n,(ans-I_simp)/(ans),'o',color='tab:red')
    print('Кол-во узлов ',n,' Погрешность ', abs((ans-I_simp)/(ans)))
pass
Simpbest_err= abs((ans-I_simp)/(ans))
Ibest_simp=I_simp
Ibest_simp
```

```python
print('Rectangaular method ', Ibest_rect,' ',Rectbest_err)
print('Trapezoidal method ', Ibest_trapz,' ',Trapzbest_err)
print('Simpson rule ', Ibest_simp,' ',Simpbest_err)
print('answer ',ans)
```
